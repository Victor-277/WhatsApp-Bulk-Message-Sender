```python
import pandas as pd
import numpy as np
from selenium import webdriver  
from selenium.common.exceptions import NoSuchElementException  
from selenium.webdriver.common.keys import Keys  
from bs4 import BeautifulSoup as soup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import time

# To run Chrome in head mode (this will open Chrome interface) 
#browser = webdriver.Chrome(executable_path=r"C:\Users\HP")  
browser = webdriver.Chrome(executable_path=r"C:\chromedriver.exe")  
#browser = webdriver.Chrome()
#browser = webdriver.Edge()
# Open WhatsApp Web
browser.get("https://web.whatsapp.com") 
print("Scan the QR code to log in.")
time.sleep(20)  # Wait for the user to scan the QR code and WhatsApp to load

# Your contacts and message
contacts = ["Alex Green", "Bella Hart", "Charlie Moon", "Dylan Brooks", "Eve Knight", "Fiona Stark", "George Webb", "Hannah Pine", "Ian Gray", "Jack Steele", "Kara Blue", "Leo Ford", "Mia Stone", "Nina Wilde"]
message = """Good afternoon! 
I hope your day is going well. Wishing you peace, joy, and success for the rest of the day!
"""


# Start sending the message to contacts
for contact in contacts:
    try:
        # Search for the contact
        search_box = browser.find_element(By.XPATH, '//div[@contenteditable="true" and @data-tab="3"]')
        search_box.click()
        search_box.send_keys(Keys.CONTROL, 'a')  # Select all text
        search_box.send_keys(Keys.BACKSPACE)  # Clear the search box
        time.sleep(1)  # Ensure the clear action is registered
        search_box.send_keys(contact)
        time.sleep(3)  # Wait for search results to load

        # Select the contact from the search results
        try:
            user = browser.find_element(By.XPATH, f'//span[@title="{contact}"]')
            user.click()
        except Exception as e:
            print(f"Could not find contact: {contact}")
            continue

        # Wait for the message box to be available
        message_box = WebDriverWait(browser, 10).until(
            EC.presence_of_element_located((By.XPATH, '//div[@contenteditable="true" and @aria-placeholder="Type a message"]'))
        )

        # Send the message
        message_box.click()
        for line in message.split("\n"):  # Send the message line by line
            message_box.send_keys(line)
            message_box.send_keys(Keys.SHIFT, Keys.ENTER)  # Add a line break
        message_box.send_keys(Keys.ENTER)  # Send the message
        time.sleep(2)  # Small delay after sending the message

        print(f"Message sent to {contact}")

    except Exception as e:
        print(f"Could not send message to {contact}. Error: {e}")
        continue

print("Messages sent successfully!")
browser.quit()
```
